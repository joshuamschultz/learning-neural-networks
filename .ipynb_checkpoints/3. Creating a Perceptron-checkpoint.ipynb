```python
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

from sklearn.datasets import make_blobs
```

When using a line to as a classifyer, we no longer use 
$$ y = m * x + b $$
But instead we solve both sides and set them to zero
This gives us two variables on one side. We can consider this as two predictors, and the 0 as a dummy variable for yes or no.
$$ w_1 x_1 + w_2 x_2 + b = y $$
In this case, we have two variables helping us classify something as "is" or "is not". By changing the wieghts of these variables, we change their impact on the outcome.

```python
X, y = make_blobs(50, 2, centers=2, random_state=2, cluster_std=2)
plt.scatter(X[:,0], X[:,1], c=y, s=30, cmap='RdBu')
```

Now instead of an x and y axis, we have an $x_1$ and $x_2$ axis
And our y is classifying whether a point is red or blue, or visually, if it is on top or bottom of our predictor line.

So we have two vectors, $X$ and $W$, in this case each two scalars long, being fed into a line equation.
$X = [x_1, x_2,....x_n]$ and $W = [w_1, w_2,... w_n]$

```python
def graph_line(W, b):
    X, Y = make_blobs(50, 2, centers=2, random_state=2, cluster_std=2)
    x = np.arange(X.min(),X.max(),1)
   
    #Calculate the intercepts
    x1_intercept = -b / W[0]
    x2_intercept = -b / W[1]
    
    slope = -(b / W[1]) / (b / W[0])
    
    y = slope * x + x2_intercept
    
    plt.scatter(X[:,0], X[:,1], c=Y, s=30, cmap='RdBu')
    plt.plot(x, y)
```

```python
W = [.38, 0.65]
b = 3
graph_line(W, b)
```

By playing with the above inputs, you can see that 
* increasing the bias moves the line towards the negative
* increasing W[0] increases slope
* increasing W[1] decreases slope

Now that we have manually played with the weights, we can create a perceptron (or predictor) using the weights we have!

We are trying to predict y, or what side of the line a new dot is on. Let's say we have a dot which is plotted at (2,-10). This would mean that $x_1$ is 2 and $x_2$ is -10. 

Let's create a perceptron method and plug in our new point.

```python
def perceptron(X,W,b):
    inputs = np.array(X)
    weights = np.array(W)
    bias = b
    
    prediction = inputs.dot(weights) + b
    
    return print(prediction)
```

```python
perceptron([2, -10], W, b)
```

This being less than 0 (below the line) would classify this point as most likely red. Visually this makes sense.
