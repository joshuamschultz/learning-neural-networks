```python
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

from sklearn.datasets import make_blobs
```

Sometimes we aren't predicting linearly, we are instead using the line to classify. In this case we want to draw the line in a way that it separates two different items.

```python
X, y = make_blobs(50, 2, centers=2, random_state=2, cluster_std=2)
plt.scatter(X[:,0], X[:,1], c=y, s=30, cmap='RdBu')
```

Visually we can see that there is a separation between the two groups, but what is the best fit line we can create to show this?

I am going to create a function creates a line, change the inputs to see how the line moves.

```python
def graph_line(m, b):
    X, Y = make_blobs(50, 2, centers=2, random_state=2, cluster_std=2)
    x = np.arange(X.min(),X.max(),1)
    slope = m
    bias = b
    
    y = slope * x + b
    
    
    plt.scatter(X[:,0], X[:,1], c=Y, s=30, cmap='RdBu')
    plt.plot(x, y)
```

```python
graph_line(0, -4)
```
